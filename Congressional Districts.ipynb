```python
url = "https://www.house.gov/representatives"
# Each state is its own table
congress = pd.read_html(url)[0]
congress.keys()
```

```python
browser = webdriver.Chrome(executable_path=r'../chromedriver.exe')
browser.get(url)
html = browser.page_source
soup = bs(html, "html.parser")
```

```python
results = soup.find_all('caption')
results[0].text.strip()
len(results)
# length is 81, which includes 25 letters. subtract 25
states = []
for x in range(len(results)-25):
    states.append(results[x].text.strip())
    
```

```python
states
```

```python
# create DF structure to append tables to
all_df = pd.DataFrame(columns=['District', 'Name', 'Party', 'Office Room', 'Phone',
       'Committee Assignment', 'State'])
```

```python
# Create a loop to read each state into df, add state name

for x in range(len(results)-25):
    state_df = pd.read_html(url)[x]
    state_df['State'] = states[x]
    all_df = pd.concat([all_df, state_df])
    del state_df
```

```python
all_df
```

```python
clean_df = all_df.drop(columns={'Party', 'Office Room', 'Committee Assignment'})
clean_df
```

```python
tracts['StAbCD '].value_counts().sort_values(ascending=False)
```

```python
#tracts df only has 436 - 435 voting districts plus DC. Remove American Samoa, Guam, Northern Mariana Islands, Puerto Rico, and Virgin Islands
remove_list = ['American Samoa', 'Guam', 'Northern Mariana Islands', 'Puerto Rico', 'Virgin Islands']
clean_df = clean_df.loc[~clean_df['State'].isin(remove_list)]
clean_df
```

```python
# review districts to clean
clean_df['District'].value_counts()
# change At Large to 00, Delegate (DC) to 98, rest to double digit numbers
small = [x for x in range(1,10)]
clean_df['District'] = clean_df['District'].replace('At Large', '00')
clean_df['District'] = clean_df['District'].replace('Delegate', '98')
clean_df['District'] = clean_df['District'].str.replace('\D', '')
clean_df['District'] = clean_df['District'].apply('{:0>2}'.format)
clean_df['District'].value_counts()
```

```python
state_series = clean_df['State']
abbr_series = []
for row in state_series:
    abbr_series.append(us_state_abbrev[row])
clean_df['Abbrev'] = abbr_series
clean_df['DistName'] = abbr_series + clean_df["District"]
```

```python
for_merge = clean_df.drop(columns={'District', 'State', 'Abbrev'})
for_merge
```

```python
combo = pd.merge(tracts, for_merge, left_on='StAbCD ', right_on='DistName')
```
