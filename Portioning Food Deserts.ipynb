```python
import pandas as pd
import numpy as np
```

### Obtained Dataset from USDA on food access measures in 2010 and 2015
- https://www.ers.usda.gov/data-products/food-access-research-atlas/
- Converted .xlsx file to .csv in Excel

```python
all_data = pd.read_csv("DataDownload2015.csv")
```

```python
all_data
```

```python
convert_columns = range(135,147) 
def convert(column):
    name = all_data.iloc[:,column].name + 'pct'    
    all_data[name] = all_data.iloc[:, column]/all_data.iloc[:, 4]

for column in convert_columns:
    convert(column)
all_data
```

### Split files into Urban/Rural

```python
urban = all_data.loc[all_data["Urban"]==1]
rural = all_data.loc[all_data["Urban"]==0]
```

### Split data into quarters by each desert range and remove non-relevant columns from each quartered DF

```python
urban_half = urban.iloc[:,np.r_[4:8,13:16,20,31:56, 147:159]]
```

```python
urban_one = urban.iloc[:,np.r_[4:8,13:16,21,57:82, 147:159]]
```

```python
rural_ten = rural.iloc[:,np.r_[4:8,13:16,22,83:108, 147:159]]
```

```python
rural_twenty = rural.iloc[:,np.r_[4:8,13:16,23,109:134, 147:159]]
```

```python
urban_half.to_csv("urban_half.csv", index=False)
urban_one.to_csv("urban_one.csv", index=False)
rural_ten.to_csv("rural_ten.csv", index=False)
rural_twenty.to_csv("rural_twenty.csv", index=False)
```
