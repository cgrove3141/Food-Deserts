```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
rurten = pd.read_csv("rural_ten_cen.csv", index_col=None)
rurten
```

```python
X = rurten.drop('LILATracts_halfAnd10', axis=1)
X
```

```python
y = rurten['LILATracts_halfAnd10']
y.head(20)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_scaler = StandardScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
y_test_categorical
```

```python
model = Sequential()
model.add(Dense(units=110, activation='relu', input_dim=20))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train_scaled, y_train_categorical, epochs=30, shuffle=True, verbose=2)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"NN Results - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
model.save("Desert_model_Rural_10_trained.h5")
```
