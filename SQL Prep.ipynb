```python
state_abbrevs = []
for state in state_folders:
    state_abbrevs.append(state[5:7].upper())
```

```python
state_abbrevs
```

```python
folders = pd.DataFrame({"folder": state_folders,
                      "abbr": state_abbrevs})
```

```python
folders
```

```python
final = pd.merge(combined, folders, left_on="St ", right_on="abbr")
```

```python
final = final.drop(columns={"abbr"})
```

```python
final.head()
```

```python
co_code = final['StCty '].apply('{:0>5}'.format)
county = final["County"]
```

```python
county.str.lower()
```

```python
## formula = "c_"+co_code+"_"+county.lower()+"/DC10CT_C"+co_code+"_001.pdf"
url = 'https://www2.census.gov/geo/maps/dc10map/tract/'
formula = url+final['folder']+"c"+co_code+"_"+county.str.lower()+"/DC10CT_C"+co_code+"_001.pdf"
formula
```

```python
combined['url'] = formula
combined = combined.drop(columns={'CensusTract'})
combined.head()
```

```python
combined= combined.rename(columns={'Tract GeoID':'tract', "DistName": 'distname', "Phone": 'phone', "Name": 'name', "ZIP": 'zip', 'POP2010': 'population', 'LA1and20':'desert'})
```

```python
combined= combined.rename(columns={'Tract GeoID ':'tract', 'St ': 'state'})
combined.head()
```

```python
combined.to_csv("use_sql.csv", encoding="UTF-8")
```
